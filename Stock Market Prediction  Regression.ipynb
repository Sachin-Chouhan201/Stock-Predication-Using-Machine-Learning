```python
#import libaries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use("bmh")
```

```python
#read Data
a = pd.read_csv("C:\\Users\\sachi\\Downloads\\NSE-TATAGLOBAL11.csv")
a.tail(20)
```

```python
#This are the  coloums in given Data Sets
b = a[['Date', 'Open', 'High', 'Low', 'Last', 'Close', 'Total Trade Quantity']]
```

```python
#Print Some intial coloums
b.head()
```

```python
#Used To Describe the Data Set
b.describe()
```

```python
#Plot Graph with the help of mataplot Libary for Opening and Closing
plt.plot(b['Open'], b['Close'])
```

```python
#Colums used uin Traning Data Set
b.columns
```

```python
#Data is Feteched For Training and Testing here the X_train use for traning and X_test for testing
X_train, X_test, y_train, y_test=train_test_split(b[['High','Low','Last','Close']],b.Open,test_size=0.33)
```

```python
#this is our training model
model=LinearRegression()
```

```python
#testing model
model.fit(X_train,y_train)
```

```python
#Predection Model
model.predict(X_test)
```

```python
#Accurcy or efficency of the model
model.score(X_test, y_test)
```
